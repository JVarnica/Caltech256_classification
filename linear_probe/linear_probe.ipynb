```python
import os

def count_images(directory):
    image_extensions = ('.jpg', '.jpeg', '.png', '.bmp', '.gif')
    total_images = 0

    for root, dirs, files in os.walk(directory):
        for file in files:
            if file.lower().endswith(image_extensions):
                total_images += 1

    return total_images

train_dir = '/content/drive/MyDrive/caltech_proj/data/Caltech256_Split/train'
val_dir = '/content/drive/MyDrive/caltech_proj/data/Caltech256_Split/val'
test_dir = '/content/drive/MyDrive/caltech_proj/data/Caltech256_Split/test'
train_images = count_images(train_dir)
val_images = count_images(val_dir)
test_images = count_images(test_dir)

print(f"Total images in training set: {train_images}")
print(f"Total images in validation set: {val_images}")
print(f"Total images in test set: {test_images}")

# Determine batch size based on image count
# This is a simple example; you might want to adjust this based on your GPU memory
batch_size = min(64, train_images // 100)  # Ensures at least 100 batches, up to a max of 64
print(f"Using batch size: {batch_size}")
```
